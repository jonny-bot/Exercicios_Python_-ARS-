```python
def soma_numero(numero1, numero2):
    return numero1 + numero2
```

```python
def subtrair_numero(numero1, numero2):
    return numero1 - numero2
```

```python
def multiplicar_numero(numero1, numero2):
    return numero1 * numero2
```

```python
def dividir_numero(numero1, numero2):
    return numero1 / numero2
```

```python
opcoes = ['Somar Número', 'Subtrair Número', 'Multiplicar Número', 'Dividir Número']
```

```python
for contador, i in enumerate(opcoes):
    print(f'{contador + 1} - {i}')
```

```python
try:

    escolha = int(input('Digite Uma das Opções: '))

    if 1 <= escolha <= len(opcoes):

        numero1 = int(input('Digite o 1° Número: '))
        numero2 = int(input('Digite o 2° Número: '))

        if escolha == 1:
            soma_numero(numero1, numero2)

        if escolha == 2:
            subtrair_numero(numero1, numero2)

        if escolha == 3:
            multiplicar_numero(numero1, numero2)

        if escolha == 4:
            dividir_numero(numero1, numero2)

    else:
        print(f'Você Digitou {escolha}. Que não está entre 1 e 4!!!')

except ValueError:
    print('Digite uma Entrada Válida!!!')
```

# 1. Classe simples
Crie uma classe Carro com os atributos:

- marca, modelo, ano

E um método descricao() que retorna uma 'string' com essas informações.
Depois, instancie dois objetos e chame o método.

```python
class Carro:
    def __init__(self, marca, modelo, ano):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano

    def descricao(self):
        return f"{self.marca} {self.modelo}, ano {self.ano}"


c1 = Carro("Toyota", "Corolla", 2020)
c2 = Carro("Honda", "Civic", 2019)

print("=== Exercício 1 ===")
print(c1.descricao())
print(c2.descricao())
```

# 2. Herança

Crie uma classe Animal com os atributos:

- nome, idade

E um método falar().
Depois, crie duas subclasses:

- Cachorro → sobrescreve falar() para retornar "Au au!"

- Gato → sobrescreve falar() para retornar "Miau!"

Instancie objetos das duas subclasses e chame falar().

```python
class Animal:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def falar(self):
        return "Som genérico de animal"


class Cachorro(Animal):
    def falar(self):
        return "Au au!"


class Gato(Animal):
    def falar(self):
        return "Miau!"


dog = Cachorro("Rex", 5)
cat = Gato("Mimi", 3)

print("=== Exercício 2 ===")
print(dog.falar())
print(cat.falar())
```

# 3. Encapsulamento
Crie uma classe ContaBancaria com:

- atributo privado __saldo

- métodos depositar(valor) e sacar(valor)

- método ver_saldo() que retorna o saldo atual

Teste criando uma conta, depositando e sacando valores.

```python
class ContaBancaria:
    def __init__(self, saldo_inicial=0):
        self.__saldo = saldo_inicial

    def depositar(self, valor):
        self.__saldo += valor

    def sacar(self, valor):
        if valor <= self.__saldo:
            self.__saldo -= valor
        else:
            print("Saldo insuficiente!")

    def ver_saldo(self):
        return self.__saldo


conta = ContaBancaria(100)
conta.depositar(50)
conta.sacar(30)

print("=== Exercício 3 ===")
print("Saldo atual:", conta.ver_saldo())
```

# 4. Polimorfismo
Crie uma classe Forma com um método area().
Depois, crie subclasses:

- Quadrado (com atributo lado)

- Circulo (com atributo raio)

Cada uma deve implementar area() de forma diferente.
Instancie objetos e chame area() para ver o polimorfismo em ação.

```python
import math


class Forma:
    def area(self):
        pass


class Quadrado(Forma):
    def __init__(self, lado):
        self.lado = lado

    def area(self):
        return self.lado ** 2


class Circulo(Forma):
    def __init__(self, raio):
        self.raio = raio

    def area(self):
        return math.pi * (self.raio ** 2)


q = Quadrado(4)
c = Circulo(3)

print("=== Exercício 4 ===")
print("Área do quadrado:", q.area())
print("Área do círculo:", c.area())
```

# 5. Projeto prático: Biblioteca
Crie um sistema simples de biblioteca:

- Classe Livro com atributos titulo, autor, ano

- Classe Biblioteca que mantém uma lista de livros e tem métodos:

    °adicionar_livro(livro)

    °listar_livros()

Instancie alguns livros e adicione-os à biblioteca, depois liste todos.

```python
class Livro:
    def __init__(self, titulo, autor, ano):
        self.titulo = titulo
        self.autor = autor
        self.ano = ano

    def __str__(self):
        return f"{self.titulo} ({self.ano}) - {self.autor}"


class Biblioteca:
    def __init__(self):
        self.livros = []

    def adicionar_livro(self, livro):
        self.livros.append(livro)

    def listar_livros(self):
        for livro in self.livros:
            print(livro)


l1 = Livro("Dom Casmurro", "Machado de Assis", 1899)
l2 = Livro("O Hobbit", "J.R.R. Tolkien", 1937)

bib = Biblioteca()
bib.adicionar_livro(l1)
bib.adicionar_livro(l2)

print("=== Exercício 5 ===")
bib.listar_livros()
```

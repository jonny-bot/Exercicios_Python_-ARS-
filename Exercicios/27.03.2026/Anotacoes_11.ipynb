# Validação Cruzada, Overfitting e Numpy
Explicação do conceito de validação cruzada para avaliar o desempenho do modelo. Identificação e prevenção do overfitting em modelos de aprendizado de máquina. Utilização de Numpy para suporte às operações matemáticas e manipulação de dados.

**Validação Cruzada**

- A validação cruzada é usada para avaliar a performance de modelos de aprendizado de máquina e estimar seu comportamento em dados não vistos.
- Divide o conjunto de dados em várias partes chamadas de "dobras" ou "folds".
- Realiza múltiplas iterações de treinamento e teste, usando um conjunto diferente como teste a cada vez.
- Permite que todas as instâncias do conjunto de dados sejam utilizadas para treinamento e teste.
- Ajuda a obter uma estimativa confiável do desempenho do modelo e evita uma única divisão arbitrária dos dados.
- A validação cruzada k-fold é uma técnica comum, dividindo o conjunto de dados em k folds de tamanho aproximadamente igual.
- Treina o modelo k vezes, usando k-1 folds como treinamento e 1 fold como teste em cada iteração.
- Calcula a métrica de desempenho em cada iteração e faz uma média dos resultados no final.
- É fundamental em Data Science para avaliação robusta do desempenho do modelo, evitar overfitting e underfitting, e ajustar os hiperparâmetros.

**Overfitting e underfitting**

Overfitting e underfitting são problemas comuns em modelos de aprendizado de máquina que ocorrem quando o modelo não é capaz de generalizar bem os padrões nos dados.

- **Overfitting:** Ocorre quando o modelo se ajusta excessivamente aos dados de treinamento, decorando os padrões específicos dos dados, mas falha em generalizar para novos dados. Isso resulta em baixa capacidade de generalização e desempenho ruim em dados não vistos. Sinais de overfitting incluem desempenho muito bom nos dados de treinamento, mas desempenho pior em dados de teste ou validação.

- **Underfitting:** Ocorre quando o modelo é muito simples para capturar os padrões dos dados de treinamento. Ele falha em se ajustar adequadamente aos dados e apresenta um desempenho ruim tanto nos dados de treinamento quanto nos dados de teste ou validação. Isso geralmente ocorre quando o modelo é muito básico ou não possui capacidade suficiente para representar a complexidade dos dados.

```python
# Importação das bibliotecas necessárias
import numpy as np  # Biblioteca para operações numéricas
from sklearn.linear_model import LinearRegression  # Modelo de regressão linear
from sklearn.preprocessing import PolynomialFeatures  # Para criar features polinomiais (modelo mais complexo)
from sklearn.metrics import mean_squared_error  # Métrica de erro (MSE)

# Dados de treinamento e teste (simples, apenas para exemplo)
dados_treinamento = np.array([1,2,3,4,5])
dados_teste = np.array([6,7,8])

# Criação dos modelos
modelo_simples = LinearRegression()  # Modelo linear simples
modelo_complexo = PolynomialFeatures(degree=4)  # Transformação polinomial de grau 4 (modelo mais complexo)

# Função para calcular o erro quadrático médio (MSE)
def calcular_mse(y_real, y_pred):
  return mean_squared_error(y_real, y_pred)

# Número de folds (divisões dos dados para validação cruzada)
k = 3

# Listas para armazenar os erros de cada modelo
mse_treinamento_simples = []
mse_teste_simples = []
mse_treinamento_complexo = []
mse_teste_complexo = []

# Loop de validação cruzada
for i in range(k):
  
  # Divide os dados em "folds"
  # Pega elementos intercalados para simular divisão em partes
  fold_treinamento = dados_treinamento[i::k]
  fold_teste = dados_teste[i::k]

  # Treinando o modelo simples (regressão linear)
  modelo_simples.fit(fold_treinamento.reshape(-1,1), fold_treinamento)

  # Transformando os dados para o modelo complexo (polinomial)
  fold_treinamento_complexo = modelo_complexo.fit_transform(fold_treinamento.reshape(-1,1))

  # Criando e treinando o modelo complexo (regressão linear + polinômio)
  modelo_complexo_regressao = LinearRegression()
  modelo_complexo_regressao.fit(fold_treinamento_complexo, fold_treinamento)

  # Fazendo previsões com o modelo simples
  pred_treinamento_simples = modelo_simples.predict(fold_treinamento.reshape(-1,1))
  pred_teste_simples = modelo_simples.predict(fold_teste.reshape(-1,1))

  # Fazendo previsões com o modelo complexo
  pred_treinamento_complexo = modelo_complexo_regressao.predict(fold_treinamento_complexo)
  pred_teste_complexo = modelo_complexo_regressao.predict(
      modelo_complexo.transform(fold_teste.reshape(-1,1))
  )

  # Calculando e armazenando os erros (MSE)
  mse_treinamento_simples.append(calcular_mse(fold_treinamento, pred_treinamento_simples))
  mse_teste_simples.append(calcular_mse(fold_teste, pred_teste_simples))
  mse_treinamento_complexo.append(calcular_mse(fold_treinamento, pred_treinamento_complexo))
  mse_teste_complexo.append(calcular_mse(fold_teste, pred_teste_complexo))

# Calculando a média dos erros para cada modelo
mse_medio_treinamento_simples = np.mean(mse_treinamento_simples)
mse_medio_teste_simples = np.mean(mse_teste_simples)
mse_medio_treinamento_complexo = np.mean(mse_treinamento_complexo)
mse_medio_teste_complexo = np.mean(mse_teste_complexo)

# Análise de Underfitting e Overfitting
if mse_medio_treinamento_simples < mse_medio_teste_simples:
  print("Existe um sinal de Underfitting no modelo simples.")
elif mse_medio_treinamento_complexo < mse_medio_teste_complexo:
  print("Existe um sinal de Overfitting no modelo complexo.")
else:
  print("Os modelo não apresentando sinal de Overfitting e Underfitting")
```

**Comentários**

O resultado "Existe um sinal de Underfitting no modelo simples" indica que o modelo de regressão linear simples não é capaz de capturar adequadamente os padrões e complexidades dos dados de treinamento. Ele apresenta um desempenho ruim tanto nos dados de treinamento quanto nos dados de teste, sugerindo que o modelo é muito simplificado para se ajustar aos dados. O Underfitting ocorre quando o modelo é muito básico ou não possui capacidade suficiente para representar a complexidade dos dados, resultando em uma baixa capacidade de generalização para dados não vistos.

```python
# Importação das bibliotecas necessárias
import numpy as np  # Biblioteca para manipulação de arrays e números
from sklearn.linear_model import LinearRegression  # Modelo de regressão linear
from sklearn.preprocessing import PolynomialFeatures  # Para criar features polinomiais
from sklearn.metrics import mean_squared_error  # Função para calcular erro MSE

# Geração de 20.000 números aleatórios entre 0 e 1
dados_aleatorios = np.random.rand(20000)

# Separação dos dados em treino e teste
dados_treinamento = dados_aleatorios[:5000]   # Primeiros 5000 valores para treino
dados_teste = dados_aleatorios[5000:6000]     # Próximos 1000 valores para teste

# Criação dos modelos
modelo_simples = LinearRegression()  # Modelo linear simples
modelo_complexo = PolynomialFeatures(degree=4)  # Transformação polinomial (grau 4)

# Função para calcular o erro quadrático médio (MSE)
def calcular_mse(y_real, y_pred):
  return mean_squared_error(y_real, y_pred)

# Número de folds (divisões para validação cruzada)
k = 3

# Listas para armazenar os erros de cada modelo
mse_treinamento_simples = []
mse_teste_simples = []
mse_treinamento_complexo = []
mse_teste_complexo = []

# Loop de validação cruzada
for i in range(k):
  
  # Criação dos folds pegando elementos intercalados
  fold_treinamento = dados_treinamento[i::k]
  fold_teste = dados_teste[i::k]

  # Treinamento do modelo simples
  # Usa o próprio valor como entrada (X) e saída (y)
  modelo_simples.fit(fold_treinamento.reshape(-1,1), fold_treinamento)

  # Transformação dos dados para o modelo polinomial
  fold_treinamento_complexo = modelo_complexo.fit_transform(
      fold_treinamento.reshape(-1,1)
  )

  # Criação e treinamento do modelo complexo
  modelo_complexo_regressao = LinearRegression()
  modelo_complexo_regressao.fit(fold_treinamento_complexo, fold_treinamento)

  # Previsões com o modelo simples
  pred_treinamento_simples = modelo_simples.predict(fold_treinamento.reshape(-1,1))
  pred_teste_simples = modelo_simples.predict(fold_teste.reshape(-1,1))

  # Previsões com o modelo complexo
  pred_treinamento_complexo = modelo_complexo_regressao.predict(fold_treinamento_complexo)
  pred_teste_complexo = modelo_complexo_regressao.predict(
      modelo_complexo.transform(fold_teste.reshape(-1,1))
  )

  # Cálculo e armazenamento dos erros MSE
  mse_treinamento_simples.append(calcular_mse(fold_treinamento, pred_treinamento_simples))
  mse_teste_simples.append(calcular_mse(fold_teste, pred_teste_simples))
  mse_treinamento_complexo.append(calcular_mse(fold_treinamento, pred_treinamento_complexo))
  mse_teste_complexo.append(calcular_mse(fold_teste, pred_teste_complexo))

# Cálculo da média dos erros
mse_medio_treinamento_simples = np.mean(mse_treinamento_simples)
mse_medio_teste_simples = np.mean(mse_teste_simples)
mse_medio_treinamento_complexo = np.mean(mse_treinamento_complexo)
mse_medio_teste_complexo = np.mean(mse_teste_complexo)

# Análise de Underfitting e Overfitting
if mse_medio_treinamento_simples < mse_medio_teste_simples:
  print("Existe um sinal de Underfitting no modelo simples.")
elif mse_medio_treinamento_complexo < mse_medio_teste_complexo:
  print("Existe um sinal de Overfitting no modelo complexo.")
else:
  print("Os modelo não apresentando sinal de Overfitting e Underfitting")
```

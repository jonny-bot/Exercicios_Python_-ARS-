# Clusterização com K-means
Introdução ao algoritmo de clusterização K-means. Agrupamento de dados utilizando K-means com Scikit-Learn.

**O que é Clusterização com K-means?**

A clusterização com K-means é um algoritmo que permite agrupar dados em diferentes categorias com base em suas similaridades.

O objetivo do K-means é encontrar grupos de pontos de dados que sejam semelhantes entre si e diferentes de outros grupos. Ele usa medidas de distância para determinar a proximidade entre os pontos e atribuí-los aos clusters correspondentes.

**O algoritmo K-means funciona da seguinte forma:**

1. Escolha o número de grupos desejado (K).
2. Selecione aleatoriamente K pontos como pontos centrais iniciais.
3. Atribua cada ponto de dados ao ponto central mais próximo usando a distância euclidiana.
4. Calcule novos pontos centrais para cada grupo, obtendo a média dos pontos atribuídos a cada centro.
5. Repita os passos 3 e 4 até que os pontos centrais não se movam muito ou até que um critério de parada seja alcançado.

Ao final, teremos K grupos, onde cada ponto de dados pertence a um grupo específico.

>**Conjunto de dados fictício de informações de clientes em uma loja de roupas**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** pandas, classe KMeans do módulo cluster da biblioteca scikit-learn e matplotlib.pyplot.

2. **Carregar o conjunto de dados:** criar um DataFrame do pandas chamado 'dados' que contém as informações dos clientes.

3. **Visualizar os dados:** criar um gráfico de dispersão dos dados dos clientes.

4. **Preparar os dados de treinamento:** selecionar as colunas 'idade' e 'gastos_mensais' do DataFrame 'dados' e armazenar as características em X.

5. **Criar o modelo K-means com 2 clusters:** criar uma instância da classe KMeans com 2 clusters.

6. **Treinar o modelo:** treinar o modelo K-means com os dados de treinamento X.

7. **Obter as previsões dos clusters:** obter as previsões dos clusters para as amostras de treinamento.

8. **Adicionar as previsões ao conjunto de dados original:** adicionar uma coluna chamada 'cluster' ao DataFrame 'dados' com as previsões dos clusters.

9. **Visualizar os clusters resultantes:** criar um gráfico de dispersão dos dados.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

dados = pd.DataFrame({
    'idade': [35,55,27,38,48,32,45,25,40,33,51,48,23,31,55,62],
    'gastos_mensais': [150,80,200,300,100,220,280,170,250,90,190,160,290,270,85,120]
})

plt.scatter(dados['idade'],dados['gastos_mensais'])
plt.xlabel('Idade')
plt.ylabel('Gastos Mensais')
plt.title('Dados dos Clientes')
plt.show()
```

```python
X = dados[['idade','gastos_mensais']]
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
previsoes = kmeans.predict(X)
dados['cluster'] = previsoes
dados
```

```python
plt.scatter(dados['idade'],dados['gastos_mensais'],c=dados['cluster'])
plt.xlabel('Idade')
plt.ylabel('Gastos Mensais')
plt.title('Clusterização com K-means')
plt.show()
```

- Usamos as variáveis "idade" e "gastos_mensais" dos clientes para realizar a clusterização.
- Inicialmente, visualizamos os dados originais para entender sua distribuição.
- Aplicamos o algoritmo K-means ao conjunto de dados para agrupar os clientes em dois clusters.
- Obtemos as previsões de cluster para cada cliente com base nas suas características.
- Finalmente, visualizamos os clusters resultantes, onde cada ponto é colorido de acordo com o cluster atribuído.

>**Pontuação de Jogadores**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** pandas, KMeans e matplotlib.

2. **Carregar o conjunto de dados:** criar um DataFrame chamado 'dados' usando a biblioteca pandas.

3. **Visualizar os dados:** criar um gráfico de dispersão dos dados, onde a habilidade é plotada no eixo x e a experiência é plotada no eixo y.

4. **Preparar os dados de treinamento:** selecionar as colunas 'habilidade' e 'experiencia' do DataFrame 'dados' e armazenar as características em X.

5. **Criar o modelo K-means com 3 clusters:** criar uma instância da classe KMeans com 3 clusters.

6. **Treinar o modelo:** treinar o modelo K-means utilizando os dados de treinamento X.

7. **Obter as previsões dos clusters:** obter as previsões dos clusters para as amostras de treinamento.

8. **Adicionar as previsões ao conjunto de dados original:** adicionar uma nova coluna chamada 'cluster' ao DataFrame 'dados', contendo as previsões dos clusters.

9. **Visualizar os clusters resultantes:** cria um gráfico de dispersão dos dados.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

dados = pd.DataFrame({
    'habilidade':[8,6,9,5,7,6,8,7,9,6,5,4,8,7,6],
    'experiencia':[4,2,5,3,4,3,5,4,5,3,2,1,5,4,3]
})

plt.scatter(dados['habilidade'],dados['experiencia'])
plt.xlabel('Habilidade')
plt.ylabel('Experiência')
plt.title('Pontuação dos Jogadores')
plt.show()
```

```python
X = dados[['habilidade','experiencia']]

kmeans = KMeans(n_clusters=3)

kmeans.fit(X)

previsao = kmeans.predict(X)

dados['cluster'] = previsao

dados
```

```python
plt.scatter(dados['habilidade'],dados['experiencia'],c=dados['cluster'])
plt.xlabel('Habilidade')
plt.ylabel('Experiência')
plt.title('Clusterização com K-means')
plt.show()
```

- Utilizamos as variáveis "habilidade" e "experiencia" dos jogadores para realizar a clusterização.
- Inicialmente, visualizamos os dados originais para compreender sua distribuição.
- Aplicamos o algoritmo K-means ao conjunto de dados para agrupar os jogadores em três clusters.
- Obtivemos as previsões de cluster para cada jogador, com base nas suas características.
- Por fim, visualizamos os clusters resultantes, onde cada ponto é colorido de acordo com o cluster atribuído.

**O que é regressão linear?**

A regressão linear é um método estatístico que ajuda a encontrar uma linha ou um plano que melhor represente a relação entre variáveis, permitindo fazer previsões ou entender como elas se relacionam.

A fórmula matemática da regressão linear simples é representada por:

imagem4.webp

**y = β₀ + β₁x + ε**

onde:

*  **y** é a variável dependente que queremos estimar;
*  **x** é a variável independente;
*  **β₀** (beta) é o intercepto, que representa o valor esperado de y quando x é igual a zero;
*  **β₁** (beta) é o coeficiente de regressão, que representa a mudança esperada em y para cada unidade de mudança em x;
*  **ε** (epsilon) é o termo de erro, que representa a variação não explicada por x.

**Quais são as aplicações da regressão linear em data science?**

**Previsão e previsões:** A regressão linear é usada para fazer previsões com base em padrões históricos. Por exemplo, pode ser aplicada para prever vendas futuras com base em dados de vendas passadas, prever o preço de imóveis com base em características como tamanho, localização, etc.

**Análise de correlação:** A regressão linear permite quantificar e medir a força da relação entre variáveis. Isso ajuda a identificar e entender a correlação entre diferentes variáveis, como a relação entre o tempo de estudo e o desempenho acadêmico, a relação entre o gasto em marketing e as vendas, entre outros.

**Análise de impacto:** A regressão linear pode ser usada para analisar o impacto de uma variável independente em uma variável dependente. Por exemplo, pode ser aplicada para entender como uma mudança nos preços afeta a demanda por um produto ou como uma campanha de marketing afeta as vendas.

**Detecção de anomalias:** A regressão linear pode ser usada para identificar valores anômalos ou discrepantes nos dados. Os resíduos (diferença entre os valores observados e os valores estimados) podem ser analisados para identificar observações incomuns ou desvios do padrão esperado.

**Seleção de recursos:** A regressão linear também é usada na seleção de recursos, ajudando a identificar quais variáveis independentes são mais relevantes para prever a variável dependente. Isso é útil quando se lida com um grande número de variáveis e é necessário selecionar as mais informativas para o modelo.

**Regressão Linear Simples**

**Coeficiente de regressão:**
O coeficiente de regressão é um número que representa a relação entre as variáveis em um modelo de regressão linear. Ele indica a mudança esperada na variável dependente para cada unidade de mudança na variável independente. Por exemplo, se o coeficiente de regressão for 2, isso significa que, em média, a variável dependente aumenta em 2 unidades para cada unidade de aumento na variável independente.

**Intercepto:**
O intercepto é um valor constante em um modelo de regressão linear que representa o valor esperado da variável dependente quando a variável independente é igual a zero. Em outras palavras, é o ponto onde a linha de regressão intercepta o eixo vertical (y) quando a variável independente é zero. Se o intercepto for 0, significa que, quando a variável independente é zero, a variável dependente também é zero.

**Previsão para X:**
A previsão para X é o valor estimado da variável dependente com base no modelo de regressão linear quando fornecemos um novo valor para a variável independente (X). Em outras palavras, é a estimativa do valor da variável dependente que o modelo prevê para um determinado valor da variável independente. Por exemplo, se a previsão para X = 6 for 12, isso significa que, de acordo com o modelo, espera-se que a variável dependente tenha o valor de 12 quando a variável independente for 6.



> **Algoritmo de regressão linear simples com uma variável independente**

**Passo a passo:**

1.   **Importar as bibliotecas necessárias:** numpy e scikit-learn.
2.   **Definir os dados de exemplo:** criar um array unidimensional.
3.   **Criar o modelo de regressão linear:** criar uma instância da classe LinearRegression.
4.   **Treinar o modelo:** treinar o modelo de regressão linear com os dados de entrada X e as respostas esperadas y.
5.   **Fazer previsões para novos valores:** criar um novo array unidimensional com um valor.
6.   **Imprimir os resultados:** Coeficiente de regressão, Intercepto e Previsão para X.



```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1,2,3,4,5]).reshape((-1,1)) # Variável independente (X)
y = np.array([2,4,6,8,10]) # Variável dependente (y)

model = LinearRegression()
model.fit(X,y)

new_X = np.array([6]).reshape((-1,1))
predicted_y = model.predict(new_X)

print("Coeficiente de regressão: ", model.coef_)
print("Intercepto: ", model.intercept_)
print("Previsão para X = ", new_X[0][0], " : ", predicted_y[0])
```

**Comentários:**

O coeficiente de regressão indica o quanto a variável dependente (y) tende a aumentar para cada aumento de uma unidade na variável independente (X). Neste caso, para cada aumento de uma unidade em X, espera-se um aumento médio de 2 unidades em y. Isso mostra que há uma relação direta e constante entre X e y, onde y aumenta proporcionalmente a X.

O intercepto indica o valor esperado da variável dependente (y) quando a variável independente (X) é igual a zero. Neste caso, quando X é zero, espera-se que y também seja zero. Isso significa que a reta de regressão começa no ponto (0,0) do gráfico, indicando que não há uma influência adicional do intercepto no modelo.

A previsão para X = 6 é igual a 12.0, conforme calculado pelo modelo de regressão linear ajustado. Isso significa que, com base nos dados e no padrão estabelecido pelo modelo, quando X é igual a 6, espera-se que y seja igual a 12.0. Essa previsão é feita com base na relação estabelecida pelo coeficiente de regressão e pelo intercepto, que foram encontrados ao ajustar o modelo aos dados de exemplo.

> **Regressão linear simples com múltiplas variáveis independentes**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** numpy e scikit-learn.
2. **Definir os dados de exemplo:** criar uma matriz bidimensional contendo os valores das variáveis independentes (X).
3. **Criar o modelo de regressão linear:** criar uma instância da classe LinearRegression.
4. **Treinar o modelo:** treinar o modelo de regressão linear com os dados de entrada X e as respostas esperadas y.
5. **Fazer previsões para novos valores:** criar uma nova matriz bidimensional com os valores das variáveis independentes para fazer uma previsão.
6. **Imprimir os resultados:** Coeficiente de regressão, Intercepto e Previsão para X.


```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1,3],[2,6],[3,9],[4,12],[5,15]]) # Variável independente
y = np.array([2,4,6,8,10]) # Variável dependente

model = LinearRegression()
model.fit(X,y)

new_X = np.array([[6,18]])
predicted_y = model.predict(new_X)

print("Coeficiente de regressão: ", model.coef_)
print("Intercepto: ", model.intercept_)
print("Previsão para X =", new_X[0], " : ", predicted_y[0])
```

**Comentários:**

Para cada aumento de 1 na primeira variável, espera-se um aumento médio de 0.2 na resposta.

Para cada aumento de 1 na segunda variável, espera-se um aumento médio de 0.6 na resposta.

Quando todas as variáveis são zero, a resposta esperada é muito próxima de zero.

Com base no modelo, quando as variáveis são 6 e 18, a resposta prevista é próxima de 12.

No entanto, pode haver alguma pequena diferença devido a arredondamentos e limitações nos cálculos.

> **Regressão linear múltipla com três variáveis independentes**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** numpy e scikit-learn.
2. **Definir os dados de exemplo:** criar uma matriz bidimensional contendo os valores das três variáveis independentes (X).
3. **Criar o modelo de regressão linear:** criar uma instância da classe LinearRegression.
4. **Treinar o modelo:** treinar o modelo de regressão linear com os dados de entrada X e as respostas esperadas y.
5. **Fazer previsões para novos valores:** criar uma nova matriz bidimensional com os valores das três variáveis independentes para fazer uma previsão.
6. **Imprimir os resultados:** Coeficiente de regressão, Intercepto e Previsão para X.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1,2,3],[2,4,6],[3,6,9],[4,8,12],[5,10,15]]) # Variável independente
y = np.array([6,10,15,18,22]) # Variável dependente

model = LinearRegression()
model.fit(X,y)

new_X = np.array([[6,12,18]])
predicted_y = model.predict(new_X)

print("Coeficiente de regressão: ", model.coef_)
print("Intercepto: ", model.intercept_)
print("Previsão para X =", new_X[0], " : ", predicted_y[0])
```

**Comentários:**

Para cada aumento de 1 nas variáveis independentes, espera-se um aumento médio de 0.2857, 0.5714 e 0.8571 na resposta, respectivamente. Isso significa que as variáveis independentes têm uma influência positiva na resposta.

Quando todas as variáveis independentes são zero, a resposta esperada começa em 2. Isso é um valor fixo que contribui para a resposta, mesmo sem as variáveis independentes.

Com base no modelo, quando as variáveis independentes são 6, 12 e 18, a resposta prevista é próxima de 26. Isso significa que, de acordo com o modelo, espera-se que a resposta seja cerca de 26 quando as variáveis independentes têm esses valores.

>**Regressão linear múltipla com quatro variáveis independentes**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** numpy e scikit-learn.
2. **Definir os dados de exemplo:** criar uma matriz bidimensional contendo os valores das quatro variáveis independentes (X).

6. **Imprimir os resultados:** Coeficiente de regressão, Intercepto e Previsão para X.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1,2,3,4],[2,4,6,8],[3,6,9,12],[4,8,12,16],[5,10,15,20]]) # Variável independente
y = np.array([10,20,30,40,50]) # Variável dependente

model = LinearRegression()
model.fit(X,y)

new_X = np.array([[6,12,18,24]])
predicted_y = model.predict(new_X)

print("Coeficiente de regressão: ", model.coef_)
print("Intercepto: ", model.intercept_)
print("Previsão para X =", new_X[0], " : ", predicted_y[0])
```

Quando todas as variáveis independentes são zero, a resposta esperada também é zero. Isso significa que, sem nenhuma influência das variáveis independentes, a resposta é inicialmente zero.

Com base no modelo, quando as variáveis independentes são 6, 12, 18 e 24, a resposta prevista é 60.0. Isso indica que, de acordo com o modelo, espera-se que a resposta seja 60 quando as variáveis independentes têm esses valores.

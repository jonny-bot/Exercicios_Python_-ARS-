## Casting

Casting é o processo de conversão de uma variável de um tipo de dado para outro tipo de dado.

Em Python, existem diversos tipos de dados built-in, como inteiros (int), ponto flutuante (float), números complexos (complex), strings (str), entre outros. Às vezes, é necessário converter uma variável de um tipo de dado para outro tipo de dado para poder realizar certas operações ou para se adequar a uma determinada necessidade.

Para realizar o casting de uma variável, é necessário utilizar as funções built-in de conversão de tipo, como int(), float(), complex(), str() etc. Por exemplo, se você tem uma variável x que é do tipo inteiro e deseja convertê-la para o tipo ponto flutuante, você pode utilizar a função float(x) e atribuir o resultado à uma nova variável.

**Casting de int para float**

```python
x = 1
y = float(x)
print(y)
print(type(y))
```

**Casting de vários tipos de dados para int**

```python
x = int(1)
y = int(2.8)
z = int("3")

print(x)
print(y)
print(z)
```

**Casting de vários tipos de dados para float**

```python
x = float(1)
y = float(2.8)
z = float("3")
w = float("4.2")

print(x)
print(y)
print(z)
print(w)
```

**Casting de vários tipos de dados para string**

```python
x = str("s1")
y = str(2)
z = str(3.0)

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

## Aritmética básica
A aritmética básica é uma das principais ferramentas da programação, e o Python oferece uma maneira simples e intuitiva de realizar operações matemáticas.

A adição é feita usando o sinal de "+" e permite adicionar dois ou mais números.

Exemplo: 1 + 2 = 3

A subtração é feita usando o sinal de "-" e permite subtrair um número de outro.

Exemplo: 3 - 2 = 1

A divisão é feita usando o sinal de "/" e permite dividir um número por outro.

Exemplo: 4 / 2 = 2

A multiplicação é feita usando o sinal de "*" e permite multiplicar dois ou mais números.

Exemplo: 2 * 3 = 6.

Além disso, existe outras operações matemáticas como potenciação, raiz quadrada e módulo, que são feitas usando operadores especiais (**, sqrt(), %) respectivamente.

É importante lembrar que em python existe diferença entre divisão inteira (//) e divisão comum (/), a divisão inteira vai retornar somente o inteiro da divisão, desconsiderando a parte fracionária.

**Raiz quadrada**

A função "math.sqrt()" é usada para calcular a raiz quadrada de um número no Python. Ela é parte da biblioteca "math", que inclui várias outras funções matemáticas úteis. Para usar a função "math.sqrt()", é necessário importar a biblioteca "math" no início do código com a instrução "import math". Depois disso, a função pode ser usada para calcular a raiz quadrada de um número passado como argumento.

```python
import math
resultado = math.sqrt(9)
print(resultado)
```

É importante notar que a função só aceita números positivos como argumento e retorna sempre um número em ponto flutuante, mesmo que o resultado seja inteiro. Se o argumento passado for negativo, a função retornará uma exceção "ValueError: math domain error".

```python
import math
a = [4,2,6]
b = []

for i in a:
  b.append(math.sqrt(i))

print(b)
```

**Potenciação**

Além disso, é possível utilizar o operador ** (duas estrelas) para fazer a potenciação em vez de usar math.pow() e usando 1/2 para a raiz quadrada, ex:

```python
resultado = 9**(1/2)
print(resultado)
```

Para usar a função "pow()", é necessário importar a biblioteca "math" no início do código com a instrução "import math". Depois disso, a função pode ser usada para calcular potências passando dois argumentos, o primeiro é a base e o segundo é o expoente.

```python
import math
resultado = math.pow(2,3)
print(resultado)
```

**Ordem das operações matemáticas**

A ordem das operações matemáticas no Python segue as regras convencionais da matemática, chamadas de ordem de precedência. Isso significa que algumas operações são realizadas antes de outras. A ordem de precedência é a seguinte:

1. Operações entre parênteses: ()
2. Exponenciação: **
3. Multiplicação, divisão e resto: *, /, %
4. Adição e subtração: +, -

Por exemplo, em uma expressão como 2 + 3 * 4, a multiplicação (3 * 4) é realizada primeiro, resultando em 12. Então, a soma (2 + 3) é realizada, resultando em 14.

É importante notar que as regras de precedência podem ser modificadas usando parênteses. Por exemplo, (2 + 3) * 4 resultaria em 20, pois as operações dentro dos parênteses são realizadas primeiro.

```python
2+10*10+3
```

```python
(2+10)*(10+3)
```

## Classes
Uma classe define o comportamento de um objeto e o tipo de informação que um objeto pode armazenar. As informações em uma classe são armazenadas em atributos e as funções que pertencem a uma classe são chamadas de métodos. Uma classe filha herda os atributos e métodos de sua classe pai.

**Criando uma classe de cachorro**

```python
class Cachorro():
  # Representa um cachorro

  def __init__(self, nome):
    # Inicializa o objeto cachorro
    self.nome = nome

  def sentar(self):
    print(self.nome + " está sentado.")

# Criar uma instância da classe Cachorro com o nome "Bob"
meu_cachorro = Cachorro('Bob')

# Imprimir uma frase
print(meu_cachorro.nome + " é um ótimo cahorro.")

# Chamar um método
meu_cachorro.sentar()
```

Em Python, __init__ é um método especial chamado de construtor. Ele é usado para inicializar objetos de uma classe quando são criados. O nome __init__ é reservado e tem dois underscores (__) antes e depois do nome.

Quando um objeto é criado a partir de uma classe, o método __init__ é automaticamente chamado, permitindo que você execute qualquer inicialização necessária para o objeto. Esse método pode receber argumentos que serão passados durante a criação do objeto.

A principal finalidade do método __init__ é definir os atributos do objeto e configurar seu estado inicial. Dentro desse método, você pode atribuir valores aos atributos do objeto, definindo-os para um estado específico.

O parâmetro self é uma convenção utilizada para se referir ao próprio objeto sendo manipulado dentro de uma classe. Ele é uma referência ao objeto específico que está sendo criado ou manipulado.

No caso do método __init__, o parâmetro self é obrigatório e sempre o primeiro parâmetro de todos os métodos de uma classe. Ele representa a instância do objeto que está sendo inicializado ou utilizado.



**Herança**

```python
class CachorroBusca(Cachorro):
  # Representa um cachorro de busca

  def __init__(self,nome):
    super().__init__(nome)

  def buscar(self):
    print(self.nome + " está buscando.")

meu_cachorro = CachorroBusca('Mel')
print(meu_cachorro.nome + " é um cachorro de busca.")
meu_cachorro.sentar()
meu_cachorro.buscar()
```

## Trabalhando com arquivos

Seus programas podem ler de arquivos e escrever em arquivos. Os arquivos são abertos em modo de leitura ('r') por padrão, mas também podem ser abertos em modo de escrita ('w') e modo de adição ('a').



**Lendo um arquivo e armazenando suas linhas**

```python
from google.colab import files
arquivo_upload = files.upload()
nome_arquivo = next(iter(arquivo_upload))

with open(nome_arquivo, 'r') as arquivo:
  linhas = arquivo.readlines()

for linha in linhas:
  print(linha)
```

**Escrevendo em um arquivo**

```python
nome_arquivo = 'diario.txt'
with open(nome_arquivo, 'w') as objeto_arquivo:
  objeto_arquivo.write("Eu amo programar")
```

**Anexando a um arquivo**

```python
nome_arquivo = 'diario.txt'
with open(nome_arquivo, 'a') as objeto_arquivo:
  objeto_arquivo.write("\nEu amo criar jogos")
```

## Exceptions
As exceções ajudam você a responder adequadamente a erros que provavelmente ocorrerão. Você coloca o código que pode causar um erro no bloco try. O código que deve ser executado em resposta a um erro vai no bloco except. O código que deve ser executado somente se o bloco try for bem-sucedido vai no bloco else.

**Capturando uma exceção.**

```python
prompt = "Quantos ingressos você precisa?"
num_tickets = input(prompt)

try:
  num_tickets = int(num_tickets)
except ValueError:
  print("Por favor, tente novamente")
else:
  print("Seus ingressos estão sendo impressos")
```

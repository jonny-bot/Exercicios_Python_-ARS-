# Visualização de Dados com Matplotlib
Utilização da biblioteca Matplotlib para criar visualizações gráficas de dados. Plotagem de gráficos de linhas, barras, dispersão e histogramas.

**O que é o Matplotlib?**

O Matplotlib é uma biblioteca poderosa em Python para plotagem de gráficos, com recursos abrangentes e opções de personalização. Permite criar gráficos e visualizações de dados detalhados e de alta qualidade.

Desenvolvido por John D. Hunter e continuamente mantido pela comunidade de código aberto, o Matplotlib é gratuito e permite contribuições da comunidade. Sua natureza de código aberto oferece liberdade de uso e possibilita melhorias e correções de bugs.

Embora seja predominantemente escrito em Python, o Matplotlib utiliza trechos de código em C, Objective-C e Javascript para garantir compatibilidade com várias plataformas. Assim, pode ser usado em diferentes sistemas operacionais e ambientes de desenvolvimento.

Com uma interface flexível e documentação abrangente, o Matplotlib é amplamente adotado por cientistas de dados, engenheiros e pesquisadores para criar visualizações informativas e impressionantes. Sua versatilidade e facilidade de uso o tornam uma escolha popular para tarefas de visualização de dados em Python.

```python
import matplotlib
print(matplotlib.__version__)
```

**Gráfico de Linha**

```python
import matplotlib.pyplot as plt
anos = [2010,2011,2012,2013,2014,2015]
receita = [1000,1500,1200,1800,2000,2500]

plt.plot(anos, receita)

plt.title("Receita Anual")
plt.xlabel("Ano")
plt.ylabel("Receita (em milhões)")
plt.grid(True)

plt.show()
```

**Gráfico de Barras**

```python
import matplotlib.pyplot as plt
cidades = ['São Paulo','Rio de Janeiro','Belo Horizonte','Brasília']
populacao = [12345678, 9876543, 5432167, 8765432]

plt.bar(cidades, populacao)

plt.title("População das Cidades")
plt.xlabel("Cidades")
plt.ylabel("População")
plt.grid(True)

plt.show()
```

**Gráfico de Dispersão**

```python
import matplotlib.pyplot as plt

idade = [25,30,35,40,45,50]
salario = [3000,3500,40000,4500,5000,5500]

plt.scatter(idade, salario)

plt.title("Relação entre Idade-Salário")
plt.xlabel("Idade")
plt.ylabel("Salário")
plt.grid(True)
```

**Gráfico de pizza**

```python
import matplotlib.pyplot as plt

categorias = ['Vendas','Marketing','Financeiro','TI']
valores = [30,25,15,30]
cores = ['#FF6F00','#FFD600','#FF1744','#00B0FF']

plt.pie(valores, labels=categorias,colors=cores, autopct='%1.1f%%',startangle=90)
plt.axis('equal')

plt.title('Distribuição de Gastos')
```

**Formatação avançada de gráficos**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x, y1, label='Seno')
ax.plot(x, y2, label='Cosseno')

ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico de Seno e Cosseno')

ax.legend(loc='upper right')

ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['left'].set_linewidth(0.5)

ax.tick_params(axis='both', which='both', length=5, width=1, direction='out', color='black')

ax.annotate('Pico', xy=(np.pi/2,1), xytext=(np.pi/2,1.5),
            arrowprops=dict(facecolor='black',arrowstyle='->'))

plt.savefig('grafico_personalizado.png')

plt.show()
```

# Manipulação de Dados com Pandas
Introdução à biblioteca Pandas para manipulação e análise de dados. Carregamento, exploração e limpeza de conjuntos de dados utilizando Pandas.

**O que é o Pandas?**

O Pandas é uma biblioteca em Python amplamente utilizada para manipulação e análise de conjuntos de dados.

Ele fornece uma ampla gama de funções e ferramentas que facilitam a análise, limpeza, exploração e manipulação de dados.

O nome "Pandas" é uma combinação das palavras "Panel Data" (dados em painel) e "Python Data Analysis" (análise de dados em Python) e foi criado por Wes McKinney em 2008.

**Por que usar o Pandas?**

O Pandas é amplamente usado por cientistas de dados e analistas de dados devido às suas poderosas funcionalidades e facilidade de uso. Aqui estão os principais motivos para usar o Pandas em projetos de análise de dados:

*   **Manipulação eficiente de dados:** O Pandas oferece estruturas de dados flexíveis, como DataFrames, que permitem armazenar e manipular grandes conjuntos de dados de forma eficiente.

*   **Análise estatística avançada:** O Pandas possui uma ampla gama de funções estatísticas integradas, permitindo realizar cálculos estatísticos, como média, mediana, desvio padrão, correlação, entre outros.

*   **Limpeza de dados eficaz:** O Pandas possui recursos poderosos para lidar com dados ausentes, duplicados ou inconsistentes. Ele oferece métodos simples para filtrar, preencher ou remover valores inválidos, garantindo a qualidade dos dados para análise.

*   **Integração com outras bibliotecas:** O Pandas é frequentemente utilizado em conjunto com outras bibliotecas populares de análise de dados, como NumPy, Matplotlib e Scikit-learn, permitindo uma análise e visualização abrangentes dos dados.

*   **Flexibilidade e produtividade:** O Pandas simplifica tarefas comuns de manipulação de dados, fornecendo funcionalidades intuitivas e expressivas. Com apenas algumas linhas de código, é possível realizar operações complexas, aumentando a produtividade como cientista de dados.

**O que o Pandas pode fazer?**

O Pandas fornece respostas sobre os dados. Por exemplo:

*   Existe correlação entre duas ou mais colunas?
*   Qual é o valor médio?
*   Valor máximo?
*   Valor mínimo?

## Fundamentos da biblioteca Pandas

**Pandas Series**

```python
import pandas as pd
dados = [10,20,30,40,50]
serie = pd.Series(dados)

print(serie)

print(serie[0])

print(serie[2:4])
```

**Pandas DataFrames**

```python
import pandas as pd
dados = {'Nome': ['João','Maria','Pedro'],
         'Idade': [25,30,35],
         'Cidade': ['São Paulo','Rios de Janeiro', 'Curitiba']}
df = pd.DataFrame(dados)
print(df)

print(df['Nome'])
print(df['Idade'])

filtro = df['Idade'] > 30
print(df[filtro])
```

## Limpeza de dados

**Criando um arquivo csv para ser usado neste exemplo**

```python
# Criando um csv
dados = {
    'Nome': ['João','Maria','Carlos','Ana','Pedro','Julia','Julia'],
    'Idade': [25,30,None,35,40,45,45],
    'Cidade':['São Paulo','Rio de Janeiro','Salvador','Belo Horizonte','São Paulo',None,None],
    'Salário':[5000,6000,7000,'N/A',9000,10000,10000]
}

df = pd.DataFrame(dados)

df.to_csv('dados2.csv',index=False)
```

**Limpando células vazias**

```python
import pandas as pd
df = pd.read_csv('dados2.csv')
df_limpo = df.dropna()
print(df_limpo)
```

**Limpeza de dados em formato incorreto**

```python
import pandas as pd
import numpy as np
df = pd.read_csv('dados2.csv')
df['Salário'] = df['Salário'].replace('N/A', np.nan).astype(float).fillna(0).astype(int)
print(df)
```

**Consertando dados errados**

```python
import pandas as pd
df = pd.read_csv('dados2.csv')

df.loc[df['Idade']<30,'Idade'] = 0
print(df)
```

**Removendo duplicatas**

```python
import pandas as pd
df = pd.read_csv('dados2.csv')
df_sem_duplicatas = df.drop_duplicates()
print(df_sem_duplicatas)
```

## Recursos avançados

**Correlação de dados**

```python
import pandas as pd
df = pd.read_csv('dados2.csv')

correlacao = df.corr()
print(correlacao)
```

A matriz de correlação exibida é uma tabela que mostra as correlações entre as variáveis "Idade" e "Salário" em um conjunto de dados. Cada valor na tabela representa o coeficiente de correlação entre as duas variáveis correspondentes.

Neste caso, a matriz mostra que há uma forte correlação positiva entre as variáveis "Idade" e "Salário". O valor de 1.000000 na diagonal principal indica que a correlação entre a mesma variável é perfeita (correlação de 100%), o que é esperado. Os outros valores na tabela indicam a correlação entre pares de variáveis diferentes.

O valor de 0.997587 fora da diagonal principal indica que existe uma correlação muito alta (quase perfeita) entre a variável "Idade" e a variável "Salário". Isso sugere que, à medida que a idade aumenta, o salário também tende a aumentar de forma bastante consistente.

A matriz de correlação é uma ferramenta útil para identificar relacionamentos entre variáveis em um conjunto de dados e pode ajudar a entender melhor os padrões e tendências presentes nos dados.

**Plotando gráficos**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dados2.csv')

plt.scatter(df['Idade'], df['Salário'])
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.title('Relação entre Idade e Salário')
plt.show()
```

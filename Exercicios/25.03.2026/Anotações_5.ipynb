**O que é regressão logística?**

A regressão logística é um método usado para prever se algo pertence a uma de duas categorias possíveis. Em vez de fornecer uma resposta exata, ele calcula a probabilidade de pertencer a uma determinada categoria.

Isso é útil em situações como prever se um e-mail é spam ou não, diagnosticar uma doença com base em sintomas ou prever se um cliente cancelará um serviço.

A regressão logística usa uma fórmula matemática especial para transformar as probabilidades e encontrar os melhores coeficientes que se ajustam aos dados de treinamento.

A fórmula matemática clássica da regressão logística é:

**p = 1 / (1 + exp(-y))**

Nessa fórmula:

*  p é a probabilidade de um evento pertencer à classe de interesse.
*  exp(x) representa a função exponencial, onde x é o valor do argumento.
*  y é a combinação linear das variáveis independentes ponderadas pelos coeficientes do modelo, mais o intercepto. Essa combinação linear é calculada como: z = β0 + β1x1 + β2x2 + ... + βnxn, onde β0, β1, β2, ..., βn são os coeficientes do modelo e x1, x2, ..., xn são as variáveis independentes.

A regressão logística usa a função exponencial para transformar um valor chamado de "y" em uma probabilidade entre 0 e 1. Valores maiores de "y" aumentam a probabilidade de pertencer à classe de interesse, enquanto valores menores diminuem essa probabilidade.

Durante o treinamento, os coeficientes do modelo são ajustados para maximizar a probabilidade de observar os dados de treinamento. Esses coeficientes são usados para calcular as probabilidades de pertencer à classe de interesse para novos dados.

**Quais são as aplicações da regressão logística em data science?**

**Análise de fraudes:** A regressão logística pode ser usada para identificar transações fraudulentas com base em padrões históricos e características das transações. Por exemplo, prever se uma transação com determinadas características é fraudulenta ou não.

**Detecção de spam:** A regressão logística é eficaz na classificação de e-mails como spam ou não spam. Com base em recursos como palavras-chave, características do remetente e conteúdo do e-mail, é possível criar um modelo de regressão logística para prever a probabilidade de um e-mail ser spam.

**Diagnóstico médico:** A regressão logística é útil na área médica para auxiliar no diagnóstico de doenças. Com base em sintomas, exames e histórico do paciente, um modelo de regressão logística pode prever a probabilidade de um paciente ter uma determinada doença.

**Previsão de churn de clientes:** A regressão logística pode ser usada para prever a probabilidade de um cliente cancelar um serviço. Com base em dados históricos do cliente, como tempo de uso, padrões de consumo e interações anteriores, um modelo de regressão logística pode identificar os fatores que influenciam a probabilidade de churn e ajudar a tomar medidas preventivas.

**Análise de crédito:** A regressão logística é aplicada na análise de crédito para prever a probabilidade de inadimplência de um indivíduo com base em características como histórico de crédito, renda, histórico de pagamento e outros fatores relevantes. Essa análise ajuda a tomar decisões informadas ao conceder empréstimos ou crédito aos clientes.

**Qual o papel da regressão logística como uma técnica de classificação?**

*  A regressão logística é uma técnica de classificação binária amplamente utilizada em data science.
*  Seu objetivo é atribuir observações a uma das duas classes possíveis.
*  O algoritmo de regressão logística estima os coeficientes que melhor se ajustam aos dados de treinamento, modelando a relação entre as variáveis independentes e a variável dependente.
*  Com base nos coeficientes estimados, é possível calcular a probabilidade de uma observação pertencer a uma classe específica.
*  Um threshold (ponto de corte) é definido para classificar as observações em uma das classes com base nas probabilidades calculadas.
*  A regressão logística é aplicada em várias áreas, como medicina, finanças, marketing e análise de dados em geral.
*  É uma técnica interpretável, pois os coeficientes fornecem informações sobre a importância e o impacto das características no processo de classificação.
Permite tomar decisões informadas com base na classificação das observações.

>**Previsão de admissão na universidade**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** numpy e scikit-learn.
2. **Definir os dados de treinamento:** notas (matriz bidimensional contendo as notas dos alunos em diferentes exames) e admissao (admitido (1) ou não (0))
3. **Criar o modelo de regressão logística:** criar uma instância da classe LogisticRegression
4. **Treinar o modelo:** treinar o modelo de regressão logística com os dados de treinamento.
5. **Definir os dados de teste:** notas_teste será uma matriz bidimensional contendo as notas dos alunos que serão usadas para fazer previsões.
6. **Fazer previsões para os dados de teste:** fazer a previsão das classes (admitido ou não admitido) correspondentes aos dados de teste usando o modelo treinado.
7. **Imprimir as previsões:** imprimir a mensagem informando que o aluno com as notas especificadas será o não admitido na universidade.

```python
# Importa a biblioteca NumPy para manipulação de arrays numéricos
import numpy as np

# Importa a classe LogisticRegression da biblioteca scikit-learn
from sklearn.linear_model import LogisticRegression

# Cria a matriz 'notas' com duas colunas (duas variáveis independentes: nota1 e nota2)
# Cada linha representa as notas de um aluno
notas = np.array([[7.2,8.5],[6.5,6.9],[8.1,7.7],[5.8,5.3],[9.2,9.7]])

# Cria o vetor 'admissao' com os rótulos (variável dependente)
# 1 = admitido, 0 = não admitido
admissao = np.array([1,0,1,0,1])

# Instancia o modelo de regressão logística
modelo = LogisticRegression()

# Treina (ajusta) o modelo com os dados de notas e admissões
modelo.fit(notas, admissao)

# Cria novos dados de teste (notas de alunos que queremos prever)
notas_teste = np.array([[8.5,8.9],[6.1,6.5],[9.3,8.7]])

# Usa o modelo treinado para prever se cada aluno será admitido (1) ou não (0)
previsoes = modelo.predict(notas_teste)

# Percorre cada previsão e imprime o resultado correspondente
for i in range(len(previsoes)):
  if previsoes[i] == 1:
    print("O aluno com notas",notas_teste[i],"será admitido na universidade.")
  else:
    print("O aluno com notas",notas_teste[i],"não será admitido na universidade.")
```

>**Previsão de compra de um produto**

**Passo a passo:**

1. **Importar as bibliotecas necessárias:** numpy e scikit-learn.

2. **Definir os dados de treinamento:** idades (matriz bidimensional contendo as idades das pessoas) e compra (array unidimensional contendo os rótulos de classe correspondentes à compra do produto) onde comprou o produto (1) ou não (0).

3. **Criar o modelo de regressão logística:** criar uma instância da classe LogisticRegression.

4. **Treinar o modelo:** treinar o modelo de regressão logística com os dados de treinamento.

5. **Definir os dados de teste:** criar idades_teste, uma matriz bidimensional contendo as idades das pessoas para as quais queremos fazer previsões.

6. **Fazer previsões para os dados de teste:** fazer a previsão das classes (compra ou não compra) correspondentes às idades das pessoas usando o modelo treinado.

7. **Imprimir as previsões:** imprimir uma mensagem informando que uma pessoa com a idade especificada provavelmente comprará ou não o produto.

```python
# Importa a biblioteca NumPy para manipulação de arrays numéricos
import numpy as np

# Importa a classe LogisticRegression da biblioteca scikit-learn
from sklearn.linear_model import LogisticRegression

# Cria a matriz 'idades' com uma coluna (variável independente: idade)
# Cada linha representa a idade de uma pessoa
idades = np.array([[25],[35],[42],[28],[30],[48],[38]])

# Cria o vetor 'compra' com os rótulos (variável dependente)
# 1 = comprou o produto, 0 = não comprou
compra = np.array([0,1,1,0,1,1,0])

# Instancia o modelo de regressão logística
modelo = LogisticRegression()

# Treina (ajusta) o modelo com os dados de idades e compras
modelo.fit(idades,compra)

# Cria novos dados de teste (idades de pessoas que queremos prever)
idades_teste = np.array([[29],[37],[45]])

# Usa o modelo treinado para prever se cada pessoa comprará (1) ou não (0)
previsoes = modelo.predict(idades_teste)

# Percorre cada previsão e imprime o resultado correspondente
for i in range(len(previsoes)):
  if previsoes[i] == 1:
    print("Uma pessoa com a idade", idades_teste[i],"anos provavelmente comprará o produto.")
  else:
    print("Uma pessoa com a idade", idades_teste[i],"anos provavelmente não comprará o produto.")
```

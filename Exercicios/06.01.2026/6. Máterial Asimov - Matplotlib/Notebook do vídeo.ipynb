# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 9, 3, 8]
y2 = [1, 2, 1, 1, 2]


plt.scatter(x, y1, label="Relação entre X e Y1",
            color='red', linestyle='--', marker='o', s=100)
plt.scatter(x, y2, label="Relação entre X e Y2", color='gray', linestyle='-', marker='x')
plt.fill_between(x, y1, y2, color='orange')

plt.title("Gráfico de linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]

plt.pie(valores, labels=categorias, autopct="%1.1f%%")
plt.show()
```

```python
import pandas as pd
```

```python
dados = pd.read_csv("dados.csv")
```

```python
dados
```

```python
dias = dados['Date']
```

```python
petro = dados['PETR4']
itau = dados['ITUB4']
```

```python
fig = plt.figure(figsize=(12, 8))

plt.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')
plt.plot(dias, itau, label='Itaú', color='orange', linestyle='--', marker='.')

plt.title('Cotação - 2025/1')
plt.xlabel('Dias')
plt.ylabel('Valor da ação (R$)')
plt.legend()
plt.xticks(ticks=dias[::20])
plt.show()
```

```python
petro.diff()
```

```python
fig = plt.figure(figsize=(12, 8))

plt.hist(petro.diff(), label='Petro', color='green', alpha=0.5, bins=50)
plt.axvline(petro.diff().mean(), color='green', linestyle='--', linewidth=4)

plt.hist(itau.diff(), label='Itau', color='orange', alpha=0.5, bins=50)
plt.axvline(itau.diff().mean(), color='orange', linestyle='--', linewidth=4)

plt.title('Variação diária - 2025/1')
plt.xlabel('Variação (R$)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
petro.diff().mean()
```

```python
itau.diff().mean()
```

```python
plt.style.use('dark_background')

fig, (ax_cima, ax_baixo) = plt.subplots(figsize=(12, 8), nrows=2, sharex=True)

ax_cima.hist(petro.diff(), label='Petro', color='green', bins=50)
ax_cima.axvline(petro.diff().mean(), color='gray', linestyle='--', linewidth=4)
ax_cima.set_xlabel('Variação (R$)')
ax_cima.set_ylabel('Frequência')
ax_cima.grid('on')

ax_baixo.hist(itau.diff(), label='Itau', color='orange', bins=50)
ax_baixo.axvline(itau.diff().mean(), color='gray', linestyle='--', linewidth=4)
ax_baixo.set_xlabel('Variação (R$)')
ax_baixo.set_ylabel('Frequência')
ax_baixo.grid('on')


fig.suptitle('Variação diária - 2025/1')
fig.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))

grid = fig.add_gridspec(2, 2)

ax_linha = fig.add_subplot(grid[0, :])
ax_hist_1 = fig.add_subplot(grid[1, 0])
ax_hist_2 = fig.add_subplot(grid[1, 1])

ax_linha.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')
ax_linha.plot(dias, itau, label='Itaú', color='orange', linestyle='--', marker='.')

ax_linha.set_title('Cotação - 2025/1')
ax_linha.set_xlabel('Dias')
ax_linha.set_ylabel('Valor da ação (R$)')
ax_linha.set_xticks(ticks=dias[::20])

ax_hist_1.hist(petro.diff(), color='green', bins=50)
ax_hist_1.axvline(petro.diff().mean(), color='gray', linestyle='--', linewidth=4)
ax_hist_1.set_xlabel('Variação (R$)')
ax_hist_1.set_ylabel('Frequência')
ax_hist_1.grid('on')

ax_hist_2.hist(itau.diff(), color='orange', bins=50)
ax_hist_2.axvline(itau.diff().mean(), color='gray', linestyle='--', linewidth=4)
ax_hist_2.set_xlabel('Variação (R$)')
ax_hist_2.set_ylabel('Frequência')
ax_hist_2.grid('on')

fig.suptitle('Análise de ações')
fig.legend()
fig.tight_layout()

plt.savefig('figura.png')

plt.show()
```

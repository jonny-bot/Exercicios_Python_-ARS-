## Primeiro Gráfico

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 9, 3, 8]

plt.plot(x, y)
plt.show()
```

## Título, rótulos e legendas

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 9, 3, 8]

plt.plot(x, y, label='Relação entre X e Y')

plt.title('Gráfico de linha')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()

plt.show()
```

## Estilizando o gráfico

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 9, 3, 8]

plt.plot(x, y, label='Relação entre X e Y', color='red', linestyle='--', marker='o')

plt.title('Gráfico de linha')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()

plt.show()
```

## Múltiplas séries

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 9, 3, 8]
y2 = [1, 2, 1, 1, 2]

plt.plot(x, y1, label='Relação entre X1 e Y1', color='red', linestyle='--', marker='o')
plt.plot(x, y2, label='Relação entre X2 e Y2', color='gray', linestyle='-', marker='x')

plt.title('Gráfico de linha')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()

plt.show()
```

## Gráfico de dispersão

É só trocar `plot` por `scatter`. Além disso, podemos mudar o tamanho e o marcador:

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 9, 3, 8]
y2 = [1, 2, 1, 1, 2]

plt.scatter(x, y1, label='Relação entre X1 e Y1', color='red', linestyle='--', marker='o', s=50)
plt.scatter(x, y2, label='Relação entre X2 e Y2', color='gray', linestyle='-', marker='*', s=100)

plt.title('Gráfico de dispersão')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()

plt.show()
```

## Gráfico de Preenchimento

Dá para preencher uma região qualquer do gráfico com uma cor. Aqui, vamos preencher entre as linhas:

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 9, 3, 8]
y2 = [1, 2, 1, 1, 2]

plt.scatter(x, y1, label='Relação entre X1 e Y1', color='red', linestyle='--', marker='o', s=50)
plt.scatter(x, y2, label='Relação entre X2 e Y2', color='gray', linestyle='-', marker='*', s=100)
plt.fill_between(x, y1, y2, color='#438f57')

plt.title('Gráfico de dispersão preenchido')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()

plt.show()
```

Bônus: escolher cor hexadecimal a partir do color picker do Google!

## Gráficos de barra

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]

plt.bar(categorias, valores)
plt.show()
```

Usar `autopct` para determinar casas decimais:

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]

plt.pie(valores, labels=categorias, autopct='%1.1f%%')
plt.show()
```

## Histograma

Calculando a diferença diária no preço da ação:

```python
# Série de valores
petro
```

```python
# Calcular a diferença de um dia para o outro
petro.diff()
```

Podemos plotar este tipo de dado em um histograma:

```python
fig = plt.figure(figsize=(12, 8))

plt.hist(petro.diff(), label='Petro', color='green')
plt.hist(itau.diff(), label='Itaú', color='orange')

plt.title('Variação diária - 2025/1')
plt.xlabel('Variação (R$)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

## Modificando parâmetros

Vamos adicionar transparência e diminuir a largura de cada barra do histograma:

```python
fig = plt.figure(figsize=(12, 8))

plt.hist(petro.diff(), label='Petro', color='green', bins=50, alpha=0.5)
plt.hist(itau.diff(), label='Itaú', color='orange', bins=50, alpha=0.5)

plt.title('Variação diária - 2025/1')
plt.xlabel('Variação (R$)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

## Plotando linhas em pontos específicos

```python
petro.diff().mean()
```

```python
itau.diff().mean()
```

```python
fig = plt.figure(figsize=(12, 8))

plt.hist(petro.diff(), label='Petro', color='green', bins=50, alpha=0.5)
plt.axvline(petro.diff().mean(), color='green', linestyle='--', linewidth=4)

plt.hist(itau.diff(), label='Itaú', color='orange', bins=50, alpha=0.5)
plt.axvline(itau.diff().mean(), color='orange', linestyle='--', linewidth=4)

plt.title('Variação diária - 2025/1')
plt.xlabel('Variação (R$)')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

## Múltiplos gráficos em uma figura

Com `plt.subplots`, especificamos quantos gráficos gerar e em que posição:

```python
fig, axes = plt.subplots(figsize=(12, 8), nrows=2)
```

```python
fig, axes = plt.subplots(figsize=(12, 8), nrows=2, ncols=3)
```

```python
fig, (ax_cima, ax_baixo) = plt.subplots(figsize=(12, 8), nrows=2)
```

Para plotar em cada gráfico, fica mais fácil pegar cada um dos gráficos em uma variável. Vamos também mudar os parâmetros dos gráficos.

**Importante**: agora não usamos mais o `plt` para plotar, mas diretamente o `ax` correto! Com isso, algumas funções vão mudar de nome.

```python
fig, (ax_cima, ax_baixo) = plt.subplots(figsize=(12, 8), nrows=2)

ax_cima.hist(petro.diff(), label='Petro', color='green', bins=50)
ax_cima.axvline(petro.diff().mean(), color='black', linestyle='--', linewidth=4)
ax_cima.set_xlabel('Variação (R$)')
ax_cima.set_ylabel('Frequência')

ax_baixo.hist(itau.diff(), label='Itaú', color='orange', bins=50)
ax_baixo.axvline(itau.diff().mean(), color='black', linestyle='--', linewidth=4)
ax_baixo.set_xlabel('Variação (R$)')
ax_baixo.set_ylabel('Frequência')

fig.suptitle('Variação diária - 2025/1')
fig.legend()
plt.show()
```

## Alinhando os eixos X

## Plotando diversos gráficos em posições customizadas

Vamos gerar um gráfico com todas as informações com que trabalhamos até aqui.

Cada gráfico (`ax`) fica em uma porção da figura!

```python
import matplotlib.pyplot as plt

plt.style.use('dark_background')

fig = plt.figure(figsize=(16, 8))

# Gerando um espaço 2x2 na figura
grid = fig.add_gridspec(2, 2)

ax_linha = fig.add_subplot(grid[0, :])  # Primeira linha
ax_hist_1 = fig.add_subplot(grid[1, 0])  # Segunda linha à esquerda
ax_hist_2 = fig.add_subplot(grid[1, 1])  # Segunda linha à direita

ax_linha.plot(dias, petro, color='green', linestyle='--', marker='.')
ax_linha.plot(dias, itau, color='orange', linestyle='--', marker='.')
ax_linha.set_title('Cotação - 2025/1')
ax_linha.set_xlabel('Dias')
ax_linha.set_ylabel('Valor da ação (R$)')
ax_linha.set_xticks(ticks=dias[::20])

ax_hist_1.hist(petro.diff(), label='Petro', color='green', bins=50)
ax_hist_1.axvline(petro.diff().mean(), color='red', linestyle='--', linewidth=2)
ax_hist_1.set_title('Variação diária - Petro')
ax_hist_1.set_xlabel('Variação (R$)')
ax_hist_1.set_ylabel('Frequência')

ax_hist_2.hist(itau.diff(), label='Itaú', color='orange', bins=50)
ax_hist_2.axvline(itau.diff().mean(), color='red', linestyle='--', linewidth=2)
ax_hist_2.set_title('Variação diária - Itaú')
ax_hist_2.set_xlabel('Variação (R$)')
ax_hist_2.set_ylabel('Frequência')

fig.suptitle('Análise de ações - 2025/1')
fig.legend()
plt.show()
```

## Salvando a figura

# Código final

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('bmh')  # Ativar linhas finas
plt.style.use('dark_background')

# Lendo os dados
dados = pd.read_csv('dados.csv')
dias = dados['Date']
petro = dados['PETR4']
itau = dados['ITUB4']

# Criando a figura e os gráficos
fig = plt.figure(figsize=(16, 8))
grid = fig.add_gridspec(2, 2)
ax_linha = fig.add_subplot(grid[0, :])  # Primeira linha
ax_hist_1 = fig.add_subplot(grid[1, 0])  # Segunda linha à esquerda
ax_hist_2 = fig.add_subplot(grid[1, 1])  # Segunda linha à direita

# Gráfico da primeira linha
ax_linha.plot(dias, petro, color='green', linestyle='--', marker='.')
ax_linha.plot(dias, itau, color='orange', linestyle='--', marker='.')
ax_linha.set_title('Cotação - 2025/1')
ax_linha.set_xlabel('Dias')
ax_linha.set_ylabel('Valor da ação (R$)')
ax_linha.set_xticks(ticks=dias[::20])

# Gráfico da segunda linha à esquerda
ax_hist_1.hist(petro.diff(), label='Petro', color='green', bins=50)
ax_hist_1.axvline(petro.diff().mean(), color='red', linestyle='--', linewidth=2)
ax_hist_1.set_title('Variação diária - Petro')
ax_hist_1.set_xlabel('Variação (R$)')
ax_hist_1.set_ylabel('Frequência')

# Gráfico da segunda linha à direita
ax_hist_2.hist(itau.diff(), label='Itaú', color='orange', bins=50)
ax_hist_2.axvline(itau.diff().mean(), color='red', linestyle='--', linewidth=2)
ax_hist_2.set_title('Variação diária - Itaú')
ax_hist_2.set_xlabel('Variação (R$)')
ax_hist_2.set_ylabel('Frequência')

# Configuração final
fig.suptitle('Análise de ações - 2025/1')
fig.legend()
plt.tight_layout()

# Salvando e exibindo
plt.savefig('gráfico.png')
plt.savefig('gráfico.pdf')
plt.savefig('gráfico.svg')
plt.show()
```

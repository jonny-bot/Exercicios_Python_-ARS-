```python
import json

# dados/aula 04/dados.json
# Abrir arquivo JSON
with open("dados/aula 04/dados.json", "r", encoding="utf-8") as f:
    dados = json.load(f)

# Exibir conteúdo
print(dados)

for i in dados:
    x = i.get('X')
    y = i.get('y', i.get('Y'))
    print(f'{x} - {y}')
```
